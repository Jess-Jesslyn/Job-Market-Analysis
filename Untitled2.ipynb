```python
from langchain_community.llms import Replicate
import os
from google.colab import userdata
# Set the API token
api_token = userdata.get('api_token')
os.environ["REPLICATE_API_TOKEN"] = api_token
# Model setup
model = "ibm-granite/granite-3.3-8b-instruct"
output = Replicate(model=model,replicate_api_token=api_token)
```

```python
for i in range(5):  # Ganti jumlah sesuai kebutuhan
    position = data.loc[i, "positionName"]
    rating = data.loc[i, "rating"]
    salary = data.loc[i, "salary"]
    company = data.loc[i, "company"]

    prompt = f"""
    Berdasarkan informasi berikut, berikan penilaian apakah pekerjaan ini termasuk:
    - Sangat Menarik
    - Layak Dipertimbangkan
    - Kurang Menarik

    Berikan alasan berdasarkan rating perusahaan dan gaji.

    - Posisi: {position}
    - Perusahaan: {company}
    - Rating Perusahaan: {rating}
    - Gaji: {salary}
    """

    print(f"Prompt untuk baris {i+1}:\n{prompt}\n{'-'*80}\n")
```

```python
import time
# Buat list kosong untuk simpan hasil klasifikasi
hasil_klasifikasi = []

# Loop data dan kirim ke model
for i in range(5):  # atau range(len(data)) kalau mau semua
    position = data.loc[i, "positionName"]
    rating = data.loc[i, "rating"]
    salary = data.loc[i, "salary"]
    company = data.loc[i, "company"]

    prompt = f"""
    Berdasarkan informasi berikut, berikan penilaian apakah pekerjaan ini termasuk:
    - Sangat Menarik
    - Layak Dipertimbangkan
    - Kurang Menarik

    Berikan alasan berdasarkan rating perusahaan dan gaji.

    - Posisi: {position}
    - Perusahaan: {company}
    - Rating Perusahaan: {rating}
    - Gaji: {salary}
    """
# Kirim prompt ke model Granite
    try:
        response = output(prompt)
        hasil_klasifikasi.append(response)
        time.sleep(7)
    except Exception as e:
        hasil_klasifikasi.append(f"Error: {e}")
print(response)
```

```python
data = data.head(5)
data["klasifikasi_granite"] = hasil_klasifikasi
```

```python
data["klasifikasi_granite"].value_counts()
```

```python
# Ringkasan jumlah dan persentase
total = len(data)
summary = data["klasifikasi_granite"].value_counts(normalize=True).mul(100).round(2)
for label, persentase in summary.items():
    print(f"- {label}: {persentase}% dari total {total} pekerjaan")
```

```python
summary_prompt = f"""
Berikut ini adalah distribusi hasil klasifikasi Granite terhadap 10 pekerjaan:

{summary}

Buatkan ringkasan insight singkat dalam 2-3 kalimat yang menjelaskan pola umum, termasuk proporsi yang dominan dan alasan yang mungkin.
"""

response = output(summary_prompt)
print(response)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set theme
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))

# Bar chart
sns.barplot(
    x=summary.index,
    y=summary.values,
    palette="Set2"
)

# Tambahkan label di atas bar
for i, v in enumerate(summary.values):
    plt.text(i, v + 1, f"{v}%", ha='center', fontweight='bold')

# Judul dan label
plt.title("Distribusi Hasil Klasifikasi Granite (Dalam Persentase)", fontsize=14)
plt.xlabel("Kategori")
plt.ylabel("Persentase (%)")
plt.ylim(0, max(summary.values) + 10)  # Tambah ruang atas
plt.tight_layout()
plt.show()
```
